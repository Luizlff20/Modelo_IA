```python
import pandas as pd
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score
import pickle
```

Este trecho de código importa algumas bibliotecas essenciais para trabalhar com dados, treinamento de modelos de aprendizado de máquina e avaliação de desempenho do modelo

```python
# Carregar os dados em um DataFrame do Pandas
data = pd.read_csv('/content/pda-prouni-2017 - pda-prouni-2017 (1).csv')

```

Esta linha de código lê os dados de um arquivo CSV específico e os carrega em um DataFrame do Pandas chamado "data" para que possamos trabalhar com esses dados no código Python.

```python
# Essas colunas são selecionadas para posterior manipulação ou análise.
selected_columns = ['NOME_IES_BOLSA', 'MODALIDADE_ENSINO_BOLSA', 'NOME_CURSO_BOLSA',
                    'NOME_TURNO_CURSO_BOLSA', 'SEXO_BENEFICIARIO_BOLSA', 'RACA_BENEFICIARIO_BOLSA',
                    'REGIAO_BENEFICIARIO_BOLSA', 'SIGLA_UF_BENEFICIARIO_BOLSA', 'MUNICIPIO_BENEFICIARIO_BOLSA']

```

Esta linha de código cria uma lista chamada selected_columns que contém os nomes de colunas específicas. Cada elemento dessa lista representa o nome de uma coluna no nosso conjunto de dados. Essas colunas são selecionadas para posterior manipulação ou análise.

```python
# X contém um subconjunto do DataFrame original, incluindo apenas as colunas listadas em selected_columns.
# Enquanto y contém os dados da coluna 'TIPO_BOLSA'. Isso é comum ao realizar tarefas de aprendizado de máquina.
X = data[selected_columns].copy()
y = data['TIPO_BOLSA'].copy()
```

X = data[selected_columns].copy(): Isso está criando um novo DataFrame X selecionando apenas as colunas especificadas na lista selected_columns do DataFrame original data. A função copy() é usada para criar uma cópia dos dados, evitando assim que as alterações em X afetem o DataFrame original.

y = data['TIPO_BOLSA'].copy(): Isso está criando uma nova Série y que contém os dados da coluna 'TIPO_BOLSA' do DataFrame original data. Assim como em X, a função copy() é usada para criar uma cópia dos dados.

Portanto, X contém um subconjunto do DataFrame original, incluindo apenas as colunas listadas em selected_columns, enquanto y contém os dados da coluna 'TIPO_BOLSA'.

```python

# Essa linha de código está criando uma lista chamada categorical_columns que contém os nomes de colunas que são tratadas como variáveis categóricas.
# Em análise de dados e aprendizado de máquina, variáveis categóricas são aquelas que representam categorias ou grupos e não têm uma ordem específica associada.
categorical_columns = ['NOME_IES_BOLSA', 'MODALIDADE_ENSINO_BOLSA', 'NOME_CURSO_BOLSA',
                    'NOME_TURNO_CURSO_BOLSA', 'SEXO_BENEFICIARIO_BOLSA', 'RACA_BENEFICIARIO_BOLSA',
                    'REGIAO_BENEFICIARIO_BOLSA', 'SIGLA_UF_BENEFICIARIO_BOLSA', 'MUNICIPIO_BENEFICIARIO_BOLSA']

```

Essa linha de código está criando uma lista chamada categorical_columns que contém os nomes de colunas que são tratadas como variáveis categóricas. Em análise de dados e aprendizado de máquina, variáveis categóricas são aquelas que representam categorias ou grupos e não têm uma ordem específica associada.

As colunas listadas em categorical_columns contêm informações categóricas, como o nome da instituição de ensino (NOME_IES_BOLSA), a modalidade de ensino (MODALIDADE_ENSINO_BOLSA), o nome do curso (NOME_CURSO_BOLSA), entre outras.

```python
# Essa linha de código utiliza a função train_test_split do scikit-learn para dividir os conjuntos de dados em conjuntos de treinamento e teste.
# Em resumo, essa linha de código divide seus dados em conjuntos de treinamento e teste, permitindo que você treine um modelo de aprendizado de máquina no conjunto de treinamento e avalie seu desempenho no conjunto de teste.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, train_size=0.8, random_state=42)

```

Essa linha de código utiliza a função train_test_split do scikit-learn para dividir os conjuntos de dados em conjuntos de treinamento e teste. X e y: São os conjuntos de features (X) e rótulos (y), respectivamente.

train_test_split(): É uma função do scikit-learn que divide os dados em conjuntos de treinamento e teste.

test_size=0.2 e train_size=0.8: Esses argumentos especificam a proporção dos dados alocados para os conjuntos de teste e treinamento. Neste caso, 20% dos dados são destinados ao conjunto de teste (test_size=0.2) e 80% aos dados de treinamento (train_size=0.8).

random_state=42: Este argumento define uma semente para a geração de números aleatórios, garantindo que a divisão dos dados seja reproduzível. Se você usar a mesma semente (neste caso, 42), você obterá a mesma divisão de dados sempre que executar o código. Isso é útil para garantir a reprodutibilidade dos resultados.

X_train, X_test, y_train, y_test: São as variáveis que armazenarão os conjuntos de treinamento e teste para features e rótulos, respectivamente.

Em resumo, essa linha de código divide seus dados em conjuntos de treinamento e teste, permitindo que você treine um modelo de aprendizado de máquina no conjunto de treinamento e avalie seu desempenho no conjunto de teste.

```python
# Essa linha de código utiliza a biblioteca CatBoost para criar e treinar um modelo de classificação
# A opção cat_features=categorical_columns indica que as colunas especificadas em categorical_columns devem ser tratadas como variáveis categóricas durante o treinamento do modelo.
model = CatBoostClassifier(cat_features=categorical_columns)
model.fit(X_train, y_train)

```

Essa linha de código está utilizando a biblioteca CatBoost para criar e treinar um modelo de classificação. Vamos analisar cada parte:

CatBoostClassifier: Esta é uma classe da biblioteca CatBoost específica para problemas de classificação. CatBoost é um algoritmo de boosting que lida especialmente bem com variáveis categóricas. Este modelo é configurado para lidar automaticamente com essas variáveis categóricas, como indicado pelo argumento cat_features=categorical_columns.

cat_features=categorical_columns: Este argumento especifica quais colunas em seus dados são variáveis categóricas. A lista categorical_columns que você forneceu anteriormente é usada aqui.

model.fit(X_train, y_train): Este método treina o modelo nos dados de treinamento. X_train contém as features de treinamento e y_train contém os rótulos correspondentes. O modelo é ajustado para fazer previsões com base nestes dados.

Portanto, no final desta linha, a variável model contém um modelo treinado usando o algoritmo CatBoost.

```python
# Calcular a assertividade nos dados de teste
y_test_pred = model.predict(X_test)
assertividade = accuracy_score(y_test, y_test_pred)
print(f"Assertividade do modelo nos dados de teste: {assertividade * 100:.2f}%")
```

y_test_pred = model.predict(X_test): Utiliza o modelo treinado (model) para fazer previsões nos dados de teste (X_test). O resultado é armazenado na variável y_test_pred, que contém as previsões do modelo para os rótulos correspondentes aos dados de teste.

assertividade = accuracy_score(y_test, y_test_pred): Calcula a acurácia do modelo comparando as previsões (y_test_pred) com os rótulos reais dos dados de teste (y_test). A acurácia é uma métrica comum para modelos de classificação, representando a proporção de previsões corretas em relação ao total de previsões.

print(f"Assertividade do modelo nos dados de teste: {assertividade * 100:.2f}%"): Imprime a acurácia do modelo nos dados de teste em formato de string. O f-string é utilizado para incorporar o valor da acurácia na mensagem impressa. O resultado é apresentado como uma porcentagem com duas casas decimais.

Resumindo, essas linhas de código estão calculando a acurácia do modelo nos dados de teste e imprimindo essa métrica para avaliar o desempenho do modelo treinado. A acurácia fornece uma indicação de quão bem o modelo está realizando previsões corretas em relação ao conjunto de teste.

```python
# Exportar o modelo treinado como arquivo .pkl
with open('modelo_catboost.pkl', 'wb') as file:
    pickle.dump(model, file)
```

with open('modelo_catboost.pkl', 'wb') as file:: Abre um arquivo chamado 'modelo_catboost.pkl' no modo de escrita binária ('wb'). A cláusula with é usada para garantir que o arquivo seja devidamente fechado após a operação, mesmo se ocorrerem exceções durante o processo.

pickle.dump(model, file): Usa o módulo pickle para serializar o modelo treinado (model) e gravar os bytes serializados no arquivo aberto (file). A função dump é utilizada para esse propósito.

Ao executar essas linhas de código, o modelo treinado é salvo no arquivo 'modelo_catboost.pkl'.

```python
#Novos dados para previsão
novos_dados = {
    'NOME_IES_BOLSA': ['PONTIFÍCIA UNIVERSIDADE CATÓLICA DO PARANÁ'],
    'MODALIDADE_ENSINO_BOLSA': ['Presencial'],
    'NOME_CURSO_BOLSA': ['Farmácia'],
    'NOME_TURNO_CURSO_BOLSA': ['Matutino'],
    'SEXO_BENEFICIARIO_BOLSA': ['M'],
    'RACA_BENEFICIARIO_BOLSA': ['Branca'],
    'REGIAO_BENEFICIARIO_BOLSA': ['Sul'],
    'SIGLA_UF_BENEFICIARIO_BOLSA': ['PR'],
    'MUNICIPIO_BENEFICIARIO_BOLSA': ['CURITIBA']
}

# Criar DataFrame a partir dos novos dados
novos_dados_df = pd.DataFrame(novos_dados)

# Fazer previsões no novo conjunto de dados
novos_dados_df['PREVISAO_TIPO_BOLSA'] = model.predict(novos_dados_df)

# Calcular a assertividade nos dados de teste
y_test_pred = model.predict(X_test)
assertividade = accuracy_score(y_test, y_test_pred)
print(f"Assertividade do modelo nos dados de teste: {assertividade * 100:.2f}%")


# Exibir o DataFrame com a nova coluna de previsão
print("\nPrevisões para novos dados:")
print(novos_dados_df[['NOME_IES_BOLSA', 'PREVISAO_TIPO_BOLSA']])
```

Define um dicionário chamado novos_dados contendo informações sobre novos dados que você deseja usar para fazer previsões.

Cria um DataFrame chamado novos_dados_df a partir desse dicionário usando o Pandas.

Utiliza o modelo treinado (model) para fazer previsões no novo conjunto de dados. As previsões são adicionadas como uma nova coluna chamada 'PREVISAO_TIPO_BOLSA' no DataFrame novos_dados_df.

Calcula a assertividade do modelo nos dados de teste existentes usando a função accuracy_score.

Imprime a assertividade do modelo nos dados de teste.

Imprime o DataFrame novos_dados_df com a nova coluna de previsão, mostrando o nome da instituição de ensino ('NOME_IES_BOLSA') e a previsão do tipo de bolsa ('PREVISAO_TIPO_BOLSA').
